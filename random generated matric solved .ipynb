```python
# importing random lib to generate random matrix
import random
# assigning cells that will generate random
cells = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
#initTiles(board): this fniunction will initialize the board with a random configuration
def initTiles():
    # random generated matrix will be stored in temp
        temp = []

        for row in range(3):
            for col in range(3):
                temp.append(cells[row][col])

        for i in range(len(temp)):
            r = random.randint(0, len(temp) - 1)
            t = temp[i]
            temp[i] = temp[r]
            temp[r] = t

        for i in range(len(temp)):
            cells[i // 3][i % 3] = temp[i]
        #print("Random generated matrix")
        #print(temp)
        import numpy as np
        #lst1 = [1,3,5,7,9,11]
        m = []
        while temp != []:
            m.append(temp[:3])
            temp = temp[3:]
        # retun the random number generated matrix
        return m
# main class
class Solution:
    #moveBoard(board, tile, direction): this function will move the tile specified in the given direction
   def moveBoard(self, board):
      dict = {}
      flatten = []
      for i in range(len(board)):
         flatten += board[i]
      flatten = tuple(flatten)

      dict[flatten] = 0
    # if generated matrix is in serial then return 0
      if flatten == (0, 1, 2, 3, 4, 5, 6, 7, 8):
         return 0
        # calling fitness function
        # fitness(board): this function will give the manhattan distance from the current board to the goal board
      return self.fitness(dict)

   def fitness(self, dict):
      cnt = 0
    # append cnt for counting how many moves taken
      while True:
         current_nodes = [x for x in dict if dict[x] == cnt]
         if len(current_nodes) == 0:
            # if given matrix not possible to solve return with -1
            return -1

         for node in current_nodes:
                # solving upmove method
            next_moves = self.nextBoards(node)
            for move in next_moves:
               if move not in dict:
                  dict[move] = cnt + 1
                    # if given code solved then retrun count of moves
               if move == (0, 1, 2, 3, 4, 5, 6, 7, 8):
                    print(move)
                    return cnt + 1
         cnt += 1
        # appending no of moves

#To solve this, we will follow these steps âˆ’

#Define a function find_next() . This will take node
#moves := a map defining moves as a list corresponding to each value {0: [1, 3],1: [0, 2, 4],2: [1, 5],3: [0, 4, 6],4: [1, 3, 5, 7],5: [2, 4, 8],6: [3, 7],7: [4, 6, 8],8: [5, 7],}
#results := a new list
#pos_0 := first value of node
#for each move in moves[pos_0], do
###
#nextBoards(board): this function will return the boards which result from moving a candidate tile (using moveBoard)
   def nextBoards(self, node):
      moves = {
         0: [1, 3],
         1: [0, 2, 4],
         2: [1, 5],
         3: [0, 4, 6],
         4: [1, 3, 5, 7],
         5: [2, 4, 8],
         6: [3, 7],
         7: [4, 6, 8],
         8: [5, 7],
      }

      results = []
      pos_0 = node.index(0)
      for move in moves[pos_0]:
         new_node = list(node)
         new_node[move], new_node[pos_0] = new_node[pos_0], new_node[move]
         results.append(tuple(new_node))

      return results
ob = Solution()


# main function
def maincall():
    
    # if you  want to create Matric then disable function initTiles ()
    # and create board in matric format [[0, 1, 4], [8, 6, 7], [5, 3, 2]]
    matrix = initTiles()
    print("Random generated matric is ")
    print(matrix)
    # calling main function
    movetaken =  ob.moveBoard(matrix)
    if  movetaken == -1:
        print("gien random generated matrix cannot solvable")
    else:
        print("no of moves taken to solve puzzle ",movetaken)
#calling main function
maincall()



```
